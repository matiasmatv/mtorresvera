```python
# Importar Matplot
import matplotlib.pyplot as plt
import numpy as np

#Definir variables y valores

Days = ("10/9", "11/9", "12/9", "13/9", "14/9", "15/9", "16/9", "17/9", "18/9", "19/9", "20/9", "21/9", "22/9", "23/9", "24/9", "25/9", "26/9", "27/9", "28/9", "29/9", "30/9", "1/10", "2/10", "3/10", "4/10", "5/10", "6/10", "7/10", "8/10", "9/10", "10/10")
Pasos = (1747, 11561, 6128, 7225, 984, 6890, 18057, 4311, 8117, 2202, 14797, 590, 7205, 1215, 2406, 12435, 1542, 5305, 2063, 6459, 4141, 1206, 2365, 9938, 10557, 13458, 13122, 6687, 1404, 2982, 6161)

# Gráfico de barras
fig, ax = plt.subplots()
ax.bar(x = Days, height = Pasos, color = "#f8766d", align = "edge", width = 0.4)

# Definir las etiquetas y el título del gráfico
ax.set_ylabel('Pasos')
ax.set_xlabel('Days')
ax.set_title('Pasos al día (por cada día de septiembre y octubre)')

# plt.show()
```
